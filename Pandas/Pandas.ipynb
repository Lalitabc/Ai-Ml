## Series

```python
import pandas as pd
```

```python
a  = pd.Series([1,2,3,4,5])
print(a)
```

```python
b  = pd.Series([1,2,3,4],index=['a','b','c','d'], dtype=float,name='python')
print(b)
```

```python
s1 =pd.Series(12, index=[0,1,2,3,4,5,6,7,8])
s2 = pd.Series(12,index=[0,1,2,3,4])

print(s1+s2)
```

## Arithmetic Operation

```python
var3 = pd.DataFrame({"A":[1,2,3,4,5], "B":[11,22,33,44,55]})
var3
```

```python
var3["C"] = var3["A"]+var3["B"]
var3
```

```python
var3["C"] = var3["A"]-var3["B"]
var3
```

```python
var3["C"] = var3["A"]*var3["B"]
var3
```

```python
var3["Python"] = var3["C"]<=55
var3
```
