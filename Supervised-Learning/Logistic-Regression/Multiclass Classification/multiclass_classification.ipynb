```python
dataset = pd.read_csv('iris.csv')
```

```python
dataset['species'].unique()
```

```python
dataset['species'].replace({'Iris-setosa':'1', 'Iris-versicolor':'2', 'Iris-virginica':'3'}, inplace=True)
```

```python
dataset
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train,y_test = train_test_split(dataset[['sepal-length', 'sepal-width', 'petal-length', 'petal-width']], dataset['species'], test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.predict(x_test) #{'Iris-setosa':'1', 'Iris-versicolor':'2', 'Iris-virginica':'3'}
```

```python
x_test
```

```python
lr.score(x_test,y_test)
```
